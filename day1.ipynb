# Part 1

```python
with open("./data/day1_part_2.txt", 'r') as f:
    data = [x.split("\n")[0] for x in f.readlines()]
```

```python
coordinates = []
for elt in data:
    temp = [x for x in elt if x.isdigit()]
    if len(temp) == 1:
        coordinates.append(int(temp[0])*11)
    else:
        nums = temp[::len(temp)-1]
        nums = int(nums[0]+nums[1])
        coordinates.append(nums)
```

```python
sum(coordinates)
```

```python
mapper = {"z": [("zero", 0)],
        "o": [("one", 1)],
        "t": [("two", 2), ("three", 3)],
        "f": [("four", 4), ("five", 5)],
        "s": [("six", 6), ("seven", 7)],
        "e": [("eight", 8)],
        "n": [("nine", 9)],
    }
```

```python
coords = []

for elt in data:
    temp = []
    for k in range(len(elt)):
        if elt[k].isdigit():
            temp.append(int(elt[k]))
        else:
            if elt[k] in mapper.keys():
                possibility  = mapper[elt[k]]
                for word, num in possibility:
                    word_len = len(word)
                    if elt[k:k+word_len] == word:
                        temp.append(num)
    print(temp)
    if len(temp) == 1:
        coords.append(int(temp[0])*11)
    else: 
        nums = temp[::len(temp)-1]
        nums = int(str(nums[0])+str(nums[1]))
        coords.append(nums)
                    
            
```
